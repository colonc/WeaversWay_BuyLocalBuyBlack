```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
df_onfarm = pd.read_csv('OnFarm.csv')
df_onfarm.loc[5:55,:]
```

```python
df_onfarm.marketname.unique()
```

```python
df_onfarm.groupby('OF_ID')
df_onfarm2 = df_onfarm.copy() 
df_onfarm2.fillna(method = 'bfill', axis = 1, limit = 1)
df_onfarm2.isnull().sum()
```

```python
df_onfarm_drop = df_onfarm2.drop(columns =['createTime', 'finishTime','updateTime'])
onfarm_final = df_onfarm_drop.rename(columns = {'Location_ST':'address','Location_City':'city', 'Location_State':'state', 'Location_StateFIPS':'stateFIPS',
                                       'Location_Zip':'zip', 'marketname': 'name',
                                       'Market_Website':'website', 'Market_Facebook': 'facebook','Market_Twitter':'twitter'})
onfarm_final.head()
```

```python
onfarm_final.to_csv(r'OnFarm_Cleaned.csv')
```
