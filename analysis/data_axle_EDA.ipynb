```python
import pandas as pd
```

```python
#Loading all files into data frames for merging
df1 = pd.read_csv('CT_DC_DE.CSV')
df2 = pd.read_csv('MA_MD.CSV')
df3 = pd.read_csv('NJ.csv')
df4 = pd.read_csv('NY.csv')
df5 = pd.read_csv('PA.csv')
df6 = pd.read_csv('VA_WV.csv')
```

```python
#Creating an array of dataframes for easier manipulation
datafiles = [df1, df2, df3, df4, df5, df6]
```

```python
#Checking column names to be sure all columns have the same names before merge


for file in datafiles:
    print (file.columns.values.tolist())
    #for i in files.columns:
        #print("-", i, "\n")
    print ("\n Moving to next dataframe \n")
    print("-------------------------------------------")
```

```python
#Dropping & Renaming certain Columns

#Important to include "inplace", it updates dataframe
for file in datafiles:
    
    #Dropping Type of Business Column
    if ("Type of Business" in file.columns.values.tolist()):
        file.drop( columns = ["Type of Business"], axis = 1, inplace = True)
    
    #Renaming Columns
    file.rename(columns = {"Phone Number Combined" : "Phone Number"}, inplace = True)
    file.rename(columns = {"Phone number" : "Phone Number"}, inplace = True)
    file.rename(columns = {"Location Sales Volume Range" : "Location Sales Range"}, inplace = True)
    file.rename(columns = {"Credit Score Alpha": "Credit Score"}, inplace = True)
    file.rename(columns = {"Credit Alpha Score": "Credit Score"}, inplace = True)

    
    #Print to see of file really dropped
    print (file.columns.values.tolist())
    print("\n Moving to next file \n")
    print("----------------------------------\n")
    
    
```

```python
#All dataframes have the same column names, so we can merge o=into one big dataframe
final_df = df1.append([df2, df3, df4, df5, df6], ignore_index=True)
final_df.drop_duplicates()
final_df.head()
```
