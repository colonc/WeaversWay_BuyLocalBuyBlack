### Notes about columns

While looking through the columns, I'm thinking that some of these columns will be mapped to department names that the coop uses. Intuitively, I'm thinking about a dicionary that will show relationship between department and produts on this list. Some of the products will have same name but others will beome combined. For example, mushrooms and vegetables will be part of a category called produce. This would allow us to report out on an aggregate level the number of suppliers available for given department as well as the number of suppliers associated for given items.

```python
farmers_markets_all.groupby('FMID')
farmers_markets_all2=farmers_markets_all.copy()
farmers_markets_all3 = farmers_markets_all2.fillna(method='bfill', axis=1, limit=1)
farmers_markets_all3.isnull().sum()
# farmers_markets_all3.sort_values(by='Website') to check to see if other website values could be filled by pulling from
# another web presence - all the nans reflect entities that don't have any web presence
```

```python
#purpose of this is to find identifying information that could be used to search for duplicates in organic_operator
# just basic information right now not the product types
farmers_markets_all_basic = farmers_markets_all3[['FMID','Website','MarketName','street','city','State','zip','County','x', 'y', 'Organic', 'WildHarvested','updateTime']]
farmers_markets_all_basic.head(5)
```

```python
#exploring the MarketName column
farmers_markets_all_basic.sort_values(by="MarketName")
#MarketName appears to be a concat of Market No + Business Name
#create a new dataframe with two new columns and split the data from MarketName into the two columns
farmers_markets_all_basic2 = farmers_markets_all_basic.MarketName.str.split(" ", n=1, expand=True)

# making separate first name column from new data frame 
farmers_markets_all_basic["market"]= farmers_markets_all_basic2[0] 
  
# making separate last name column from new data frame 
farmers_markets_all_basic["name"]= farmers_markets_all_basic2[1] 
farmers_markets_all_basic.head(5)
```

```python
farmers_markets_all_basic.tail(5)
```

```python
# Column names: remove white spaces and convert to lower case
farmers_markets_all_basic.columns = farmers_markets_all_basic.columns.str.strip().str.lower()
farmers_markets_all_basic.columns
```

```python
farmers_markets_all_basic = farmers_markets_all_basic [['fmid', 'name','website','street','city','state','zip','county','x', 'y', 'organic', 'wildharvested','updatetime']]
farmers_markets_all_basic.head(5)
```
