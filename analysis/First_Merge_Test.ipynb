<h1> <b>Modifiying Website URLS</b> </h1>
    <p> Things done to clean up website column: <p>
    <li> Imported validators libraries -  for validating website urls </li>
    <li> Created a new df that contains just website column </li>
    <li> Created a function, that splits existing urls that contains specified words into two parts, and concatenated "http://www" with the split word </li>
    <li> For those that were not modified by first function, created a second function that checks if they end in specific words, and then concatened with "http://www" </li>
    <li> Used validator to check if the urls are valid </li>
    <li> Finally, assigned the validated column back to original df </li>

```python
#Taking just the website column to work on
website_df = all_operators[['website']].copy()
website_df
```

```python
#Function to modify website urls

words_to_search = ['www','http', 'https']

def edit_url(x):
    new_str = str(x)
    for i in words_to_search:
        if i.lower() in new_str.lower():
            new_word = new_str.split('.', 1)
            return "http://www." + new_word[1]
        else:
             return x
```

```python
website_df['edited'] = website_df['website'].apply(edit_url)

website_df
```

```python
my_List = ('com', 'org', 'net')
def new_edit(x):
    new_str = str(x)
    if new_str.startswith("http://www."):
        return new_str
    elif new_str.endswith(my_List):
        return "http://www." + new_str
    else:
        return x
```

```python
website_df['newly_edited'] = website_df['edited'].apply(new_edit)
```

```python
#Using Validator library to check if url is valid
def valid_url(x):
    new_str = str(x)
    if (validators.url(new_str) == True):
        return new_str
    else:
        return np.nan
```

```python
website_df['actual_website'] = website_df['newly_edited'].apply(valid_url)
```
