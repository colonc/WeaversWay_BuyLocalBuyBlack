```python
import pandas as pd 
import numpy as np
```

```python
bipoc_farms = pd.read_csv('BIPOC in Agriculture and Food_ A Resource Database - Farms & Gardens.csv')
bipoc_farms
```

```python
bipoc_farms["State"].unique()
```

```python
nearby_states = ['PA', 'NY', 'NJ', 'VA', 'MD','RI', 'DC']
filt = bipoc_farms['State'].isin(nearby_states)
bipoc_farms_condensed = bipoc_farms.loc[filt]
bipoc_farms_condensed
bipoc_renamed = bipoc_farms_condensed.rename(columns ={"Name " : "Name"})
```

```python
#remove unnecessary columns 
bipoc_farms_final = bipoc_renamed[['Name', 'Sheet Description', 'Website', 'City', 'State', 'Facebook', 
                                             'Instagram']]
bipoc_farms_final
```

```python
Shoppeblack_minorityCSA = pd.read_csv("Shoppeblack and CSA.csv")
Shoppeblack_minorityCSA
```

```python
Black_farmers_index = pd.read_csv("Black farmers index.csv", encoding = 'latin1')
Black_farmers_renamed = Black_farmers_index.rename(columns = {'State ' : 'State', 'Name ': 'Name', 'Instagram ' : 'Instagram'})
```

```python
Black_farmers_renamed.columns
```

```python
Shoppeblack_minorityCSA.columns
```

```python
#fix spaces in names 
Shoppeblack_renamed = Shoppeblack_minorityCSA.rename(columns = {'Name ' : 'Name', 'Website ' : 'Website'})
```

```python
#combine all farms 
minority_owned_farms = pd.concat([Shoppeblack_renamed, Black_farmers_renamed, bipoc_farms_final], sort = False)
minority_owned_farms.drop_duplicates(subset = 'Name', keep = 'first', inplace = True)
minority_owned_farms.duplicated(subset = 'Name').sum()
minority_owned_farms
```
