```python
from numpy import nan
import pandas as pd
```

```python
# Read dictionary from string in file
def read_dict_from_file(filename):
  # Open file as f in read-only mode
  with open(filename, "r") as f:
    # Parse content into variable
    my_string = f.read()
    # Return evaluated dictionary
    return eval(my_string)
print(type(read_dict_from_file("../data/OnFarm_data/data_axle_dict.txt")))

file = read_dict_from_file("../data/OnFarm_data/data_axle_dict.txt")
```

```python
def getList(dict): 
    return dict.keys()

dict_keys = getList(file)
dict_keys
```

```python
check = pd.read_csv("../data/OnFarm_data/all_operators.csv")
```

```python
checked_operators = check[check['name'].isin(dict_keys)]
```
