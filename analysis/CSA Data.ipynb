```python
import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
import numpy as np 
df_19002 = pd.read_csv('../data/Local_Food_Directories_CSA_Directory_Zip_19002_ Radius_200.csv')
df_19119 = pd.read_csv('../data/Local_Food_Directories_CSA_Directory_Zip_19119_ Radius_200.csv')
df_19128 = pd.read_csv('../data/Local_Food_Directories_CSA_Directory_Zip_19128_ Radius_200.csv')                                                                                                  
```

```python
df_19002.columns
```

```python
CSA_complete = pd.concat([df_19002, df_19119, df_19128])
CSA_complete.index
```

```python
#food columns disappear here 
print(CSA_complete.columns)
```

```python
#remove duplicates
CSA_complete.drop_duplicates(subset = 'CSA_ID', keep='first', inplace = True)
CSA_complete.duplicated(subset = 'CSA_ID').sum()
```

```python
CSA_complete.isnull().sum() 
```

```python
CSA_complete.groupby('CSA_ID')
CSA_complete2 = CSA_complete.copy() 
CSA_complete2.fillna(method = 'bfill', axis = 1, limit = 1)
CSA_complete2.isnull().sum()
```

```python
#remove empty and irrelevant columns
CSA_drop = CSA_complete2.drop(columns =['Mailing_ST', 'Mailing_City', 'Mailing_State', 'Mailing_StateFIPS',
       'Mailing_Zip', 'createTime', 'finishTime','updateTime'])
```

```python
#rename columns for future uniformity 
CSA_final = CSA_drop.rename(columns = {'HQ_ST':'address','HQ_City':'city', 'HQ_State':'state', 'HQ_StateFIPS':'stateFIPS',
                                       'HQ_Zip':'zip','CSA_ManagerName': 'manager_name', 'CSA_Email': 'email',
                                       'CSA_Phone':'phone', 'CSA_Website': 'website','CSA_Facebook':'facebook',
                                       'CSA_Twitter':'twitter'})
CSA_final
```

```python
CSA_simplified = CSA_final[['CSA_ID', 'MarketName', 'address', 'city', 'state', 'stateFIPS', 'zip',
       'manager_name', 'email', 'phone',]]
CSA_simplified.head(20)
```
