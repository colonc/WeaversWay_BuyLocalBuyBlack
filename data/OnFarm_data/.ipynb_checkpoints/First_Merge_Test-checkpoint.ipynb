### 1. Drop rows with states outside of the 9 associated with our 200 mile radius states.

```python
radius_states=['Pennsylvania', 'New Jersey','Delaware', 'Maryland', 'New York', 'Connecticut', 'District of Columbia', 'Virginia', 'West Virginia', 'Massachusetts']
```

```python
all_operators = all_operators.loc[all_operators.state.isin(radius_states)]
all_operators
```

### Look to see where ['other_name'] appears in [name]. 
Keep these rows because not all the columns match across and despite being the same name, they appear to be different businesses. Drop other_name column because the other databases do not have this information. In hindsight, probably shouldn't have included it in the merge.


```python
other_name = all_operators.name.unique()
```

```python
all_operators['compared'] = np.where((all_operators.other_name.isin(other_name)), 'true', 'false')
```

```python
all_operators.loc[all_operators.compared=='true']
```
